KNN

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix


df = pd.read_csv("..\\Datasets\\diagnosed_cbc_data_v4.csv")

# --- 2. Features und Zielvariable trennen ---
X = df.drop("Diagnosis", axis=1)
y = df["Diagnosis"]

# --- 3. Daten normalisieren ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 4. Trainings- und Testdaten splitten ---
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# --- 5. KNN-Modell erstellen und trainieren ---
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)

# --- 6. Vorhersage und Bewertung ---
y_pred = knn.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# --- 7. Beispielvorhersage ---
example = [[7.0, 32.0, 60.0, 2.0, 4.0, 4.0, 8.5, 30.0, 78.0, 23.0, 30.0, 150, 12.0, 0.15]]
example_scaled = scaler.transform(example)
prediction = knn.predict(example_scaled)
print("\nVorhersage für Beispiel:", prediction[0])

# precision = TP/(TP+FP)
# recall/sensitivity = TPR = True positive rate, TP/(TP+FN) -> ^TPR = good at detecting if patient is actually healty/has this disease
# f1-score = (2*precision*recall)/precision+recall -> Mean of precision and recall -> ^F1=better performance
# support = No. of values
```

Decision Tree

```python
# Decision Tree Classifier für Blutbild-Daten
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from io import StringIO

# --- 1. Daten laden ---
df = pd.read_csv("..\\Datasets\\diagnosed_cbc_data_v4.csv")


# --- 2. Features & Zielvariable ---
X = df.drop("Diagnosis", axis=1)
y = df["Diagnosis"]

# --- 3. Skalierung (optional, für Entscheidungsbaum nicht nötig, aber zur Einheitlichkeit) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 4. Train/Test-Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# --- 5. Decision Tree trainieren ---
dt = DecisionTreeClassifier(random_state=42, max_depth=3)
dt.fit(X_train, y_train)

# --- 6. Bewertung ---
y_pred = dt.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# --- 7. Visualisierung des Entscheidungsbaums ---
plt.figure(figsize=(14, 8))
plot_tree(
    dt,
    feature_names=X.columns,
    class_names=dt.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Classifier – Blutbild-Daten")
plt.show()
```

Random Forest

```python
# Random Forest Classifier für Blutbild-Daten
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from io import StringIO
import numpy as np

# --- 1. Daten laden ---
df = pd.read_csv("..\\Datasets\\diagnosed_cbc_data_v4.csv")


# --- 2. Features und Zielvariable ---
X = df.drop("Diagnosis", axis=1)
y = df["Diagnosis"]

# --- 3. Skalieren (optional für Random Forest, hier zur Einheitlichkeit) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 4. Train/Test-Split ---
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# --- 5. Random Forest trainieren ---
rf = RandomForestClassifier(
    n_estimators=100,      # Anzahl Bäume
    max_depth=5,           # maximale Tiefe (optional)
    random_state=42
)
rf.fit(X_train, y_train)

# --- 6. Bewertung ---
y_pred = rf.predict(X_test)
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# --- 7. Feature Importance Plot ---
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), X.columns[indices], rotation=45, ha='right')
plt.ylabel('Feature Importance')
plt.title('Random Forest – Wichtigste Merkmale')
plt.tight_layout()
plt.show()
```

Gaussian Mixture Model

```python
# Gaussian Mixture Model (GMM) für Blutbild-Daten
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from io import StringIO
import numpy as np

# --- 1. Daten laden ---
df = pd.read_csv("..\\Datasets\\diagnosed_cbc_data_v4.csv")


# --- 2. Features und Zielvariable ---
X = df.drop("Diagnosis", axis=1)
y = df["Diagnosis"]

# --- 3. Skalieren ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 4. Gaussian Mixture Model (3 Cluster, da wir 3 Diagnosen haben) ---
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X_scaled)
clusters = gmm.predict(X_scaled)

# --- 5. Cluster-Ergebnisse ---
df["Cluster"] = clusters
print("Cluster-Zuordnung:")
print(df[["Diagnosis", "Cluster"]])

# --- 6. Vergleich Cluster vs. Diagnosen ---
print("\nKontingenztabelle (Cluster vs Diagnosis):")
print(pd.crosstab(df["Cluster"], df["Diagnosis"]))

# --- 7. PCA für 2D-Plot ---
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', s=120, edgecolor='k')
for i, label in enumerate(y):
    plt.text(X_pca[i, 0] + 0.05, X_pca[i, 1], label.split()[0], fontsize=9)
plt.title("Gaussian Mixture Model (PCA 2D-Projektion)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Cluster")
plt.tight_layout()
plt.show()
```
