```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read Diagnosis dataset

```python
df = pd.read_csv("diagnosed_cbc_data_v4.csv", sep=",")
df.head(n=5)
```

Check all Datatypes

```python
df.dtypes
```

Check for duplicates

```python
df.duplicated().any()
```

Display all duplicate rows

```python
df[df.duplicated(keep=False)].sort_values(by=list(df.columns)).head(100)
```

```python
df.drop_duplicates(inplace=True)
df.head()
```

Get total rows after dropping all duplicates

```python
df.count
```

Check if there's empty cells

```python
df.isna().any() # add another .any() to aggregate to a single Boolean
```

Get statistical overview of data

```python
df.describe()
```

Correlation matrix

```python
plt.subplots(figsize=(12, 12))
X = df.drop('Diagnosis', axis=1)
sns.heatmap(X.corr(), annot=True, cmap='RdYlGn_r', linewidths=0.5, fmt='.2f')
```

Detail Boxplot per Disease

```python
df_iron = df[df["Diagnosis"] == "Iron deficiency anemia"]

feature_cols = df_iron.select_dtypes(include="number").columns

plt.figure(figsize=(12, 6))
df_iron[feature_cols].plot(kind='box', subplots=True, layout=(7, 2), figsize=(10, 18), sharex=False)
plt.title("Boxplots der numerischen Features â€“ Iron Deficiency Anemia")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Boxplot of feature values per Disease side by side

```python
df_iron = df[df["Diagnosis"] == "Iron deficiency anemia"]

feature_cols = df_iron.select_dtypes(include="number").columns

plt.figure(figsize=(12, 6))
df_iron[feature_cols].boxplot()
plt.title("Boxplots der numerischen Features â€“ Iron Deficiency Anemia")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
